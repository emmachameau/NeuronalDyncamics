# Spatial Working Memory

In this exercise we study a model of spatial working memory. The model has been introduced by Compte et. al. 



**Ring model with spiking neurons**



The spatial continuum model 
We focus on networks that have a spatial structure. In doing so we emphasize two characteristic features of the cerebral cortex, namely the high density of neurons and its virtually two-dimensional architecture.
Each cubic mm of cortical tissue contains more than 10^4 neurons. This suggests that an averaged population activity is more appropriate than a description on the single neuron level. Neurons will no longer be labeled by discrete indices but by continuous variables that give their spatial position on the sheet. The coupling of two neurons i and j is replaced by the average coupling strength between neurons at position x and those at position y. Consider a single sheet of densely packed neurons. We assume that all neurons has similar properties and that the connectivity is homogeneous and isotropic, i.e., that the coupling strength of two neurons is a function of their distance only. 
In the following, we will consider a connectivity pattern that is **excitatory for proximal neurons and predominantly inhibitory for distal neurons**. 



With Mexican hat coupling, two different regimes emerge:

(i) In the input-driven regime the activity pattern is, in the absence of input, spatially uniform. In other words, any spatial structure in the neuronal activity pattern is causally linked to the input.

(ii) In the regime of bump-attractors , spatially localized activity patterns develop even in the absence of input (or with spatially uniform input).

**Input driven regime**

Field models with a Mexican-hat interaction kernel generically generate contrast enhancement in the input driven regime. Because of the nonlinear lateral interactions, an incoming spatial input pattern is transformed. For example, a spatial input with rectangular profile boosts activity at the borders, while a smooth input with sinusoidal modulation across space boosts activity at the maximum.

The Mexican-hat interaction pattern has a long tradition in theoretical neuroscience, but, from a biological perspective, it has two major shortcomings. First, in field models with Mexican-hat interaction, the same presynaptic population gives rise to both excitation and inhibition whereas in cortex excitation and inhibition require separate groups of neuron (Dales law). Second, inhibition in Mexican-hat connectivity is of longer range than excitation whereas biological data suggests the opposite. In fact, inhibitory neurons are sometimes called local interneurons because they only make local interactions. Pyramidal cells, however, make long-range connections within and beyond cortical areas.

Simple and complex cells in visual cortex respond best if they are stimulated by a slowly moving grating with optimal orientation and of a size that is matched to the cells receptive field. If the grating is optimally oriented but larger than the receptive field, the response is reduced compared to that of a smaller grating. At first sight, this finding is consistent with contrast enhancement through Mexican-hat interaction: a uniform large stimulus evokes a smaller response because it generates inhibition from neurons which are further apart. Paradoxically, however, neurons receive less inhibition with the larger stimulus than with the smaller one.



- group neurons with overlapping receptive fields of similar orientation preference into a single population; they excite each other
- we record the neuron in the center; neurons with receptive fields far away from  the recorded neuron inhibit its activity
- The excitatory neurons in the central population project onto a group of local inhibitory interneurons, but also onto populations of other inhibitory neurons further apart. Each population of inhibitory neurons makes only local connections to the excitatory population in their neighborhood. Input to the central group of excitatory neurons therefore induces indirect inhibition of excitatory neurons further apart. 
- the network parameters are set such that the network is in the inhibition-stabilized regime. A network is said to be inhibition-stabilized if the positive feedback through recurrent connections within an excitatory population is strong enough to cause run-away activity in the absence of inhibition. To counterbalance the positive excitatory feedback, inhibition needs to be even stronger. As a result, an inhibition-stabilized network responds to a positive external stimulation of inhibitory neurons with a decrease of both excitatory and inhibitory activity (see Exercises).
- If the coupling from excitatory populations to neighboring inhibitory populations is stronger than that to neighboring excitatory populations, an inhibition-stabilized network can explain the phenomenon of surround suppression and at the same time account for the fact that during surround suppression both inhibitory and excitatory drive are reduced

**Bump attractors and spontaneous pattern formation**

From a computational point of view bistable systems are of particular interest because they can be used as memory units. For example a homogeneous population of neurons with all-to-all connections can exhibit a bistable behavior where either all neurons are quiescent or all neurons are firing at their maximum rate. By switching between the inactive and the active state, the neuronal population is able to represent, store, or retrieve one bit of information. The exciting question that arises now is whether a neuronal net with distance-dependent coupling can store more than just a single bit of information, but spatial patterns of activity. Sensory input, e.g., visual stimulation, could switch part of the network to its excited state whereas the unstimulated part would remain in its resting state. **Due to bistability this pattern of activity could be preserved even if the stimulation is turned off again and thus provide a neuronal correlate of working memory.**

**Sense of Orientation and Head Direction Cells**
- Each head direction cell has a different receptive field, meaning it fires when the head is pointing in a specific direction.
- Even when external cues (like lights) are removed, these cells continue to fire, indicating that the animal remembers its head orientation in the absence of sensory input.
- When the external cues are rotated and then reintroduced, the head direction representation shifts to align with the new visual cues.
- Ring Model for Head Direction:
To mathematically represent this continuous head direction variable, ring models are used.
In a ring model, neurons are arranged in a circular structure, where each neuron represents a preferred head direction.
Excitation and Inhibition Dynamics:
Neurons with similar preferred head directions excite each other.
Neurons with opposite head directions inhibit each other.
This results in a Mexican-hat connectivity pattern, a common feature in neural network models where local excitation is balanced by long-range inhibition.
- The ring model is designed as a bump attractor network, meaning that once the network forms a pattern of activity, it can persist even in the absence of external stimulation.

*A bump attractor is a type of neural network dynamics where neuronal activity forms a stable, localized region of excitation, referred to as a bump, that persists over time. These attractors are often used to model continuous variables in neural systems, such as head direction, spatial position, or working memory. They are fundamental in explaining how the brain maintains information about orientation, position, or movement, even in the absence of sensory input.*

*For the time point t=400ms, sketch the firing rate across the population (neuron index on the x-axis, per-neuron firing rate on the y-axis).*

```python
import matplotlib.pyplot as plt
import brian2 as b2
import numpy as np

spike_times = spike_monitor_excit.t / b2.ms  
neuron_indices = spike_monitor_excit.i

time_window = 400.0  

num_neurons = spike_monitor_excit.source.N
firing_rates = np.zeros(num_neurons)

for neuron_id in range(num_neurons):
    spikes_for_neuron = spike_times[neuron_indices == neuron_id]
    firing_rates[neuron_id] = np.sum(spikes_for_neuron <= time_window) / (time_window / 1000.0)  # Hz

plt.figure(figsize=(10, 6))
plt.bar(range(num_neurons), firing_rates, width=1.0, color='blue')
plt.xlabel("Neuron Index")
plt.ylabel("Firing Rate (Hz)")
plt.title(f"Firing Rate Across the Population at t = {time_window} ms")
plt.show()
```

**Mapping the neuron index onto its preferred direction**

This function maps a neuron index to its corresponding preferred orientation , assuming that preferred orientations are evenly spaced across neurons.

```python
import numpy as np

def get_orientation(idx_list, N):
    """
    maps a list of neuron indices to their preferred directions.

    parameters:
    - idx_list: List of neuron indices (subset of monitored neurons).
    - N: Total number of neurons in the excitatory population.

    returns:
    - a list of preferred directions corresponding to the given indices.
    """
    
    return [round(((idx + 0.5) / N) * 360, 2) for idx in idx_list]

# test 
print(get_orientation([0, 1, 5, 10], 11))  # Expected: [16.36, 49.09, 180.0, 343.64]
print(get_orientation([0, 1, 499, 500, 999], 1000))  # Expected: [0.18, 0.54, 179.82, 180.18, 359.82]
```

**Extracting spikes from the spike monitor**

To estimate (t), we count how many times each neuron spiked in a given time window.

The population vector  changes over time due to drift and diffusion which is why we are interested in (t). As we are dealing with spikes (discrete point events), and a small number of neurons, we have to average the population activity over some time window around t (i.e. [tmin=ttwindow/2,tmax=t+twindow/2]) to get an estimate of (t).



```python
%matplotlib inline

import brian2 as b2
from neurodynex3.working_memory_network import wm_model
from neurodynex3.tools import plot_tools
import matplotlib.pyplot as plt
import numpy as np

rate_monitor_excit, spike_monitor_excit, voltage_monitor_excit, idx_monitored_neurons_excit, rate_monitor_inhib, spike_monitor_inhib, voltage_monitor_inhib, idx_monitored_neurons_inhib, w_profile = wm_model.simulate_wm(
    stimulus_center_deg=120, stimulus_width_deg=30, stimulus_strength=.06 * b2.namp, t_stimulus_start=100 * b2.ms, t_stimulus_duration=200 * b2.ms, sim_time=500. * b2.ms)


fig, ax_raster, ax_rate, ax_voltage = plot_tools.plot_network_activity(
    rate_monitor_excit, spike_monitor_excit, voltage_monitor_excit, t_min=0. * b2.ms)
plt.show()


```

```python
def get_spike_count(spike_monitor, spike_index_list, t_min, t_max):
    nr_neurons = len(spike_index_list)
    spike_count_list = np.zeros(nr_neurons)  # Initialize counts to 0
    spike_trains = spike_monitor.spike_trains()
    print(f"type: {type(spike_trains)} value: {spike_trains}")
    
    
    for i, neuron_index in enumerate(spike_index_list): 
        spikes = spike_trains[neuron_index]
        for x in spikes:
            if x >= t_min and x < t_max:
                print(i)
                spike_count_list[i] = x
            else:
                #print(f"not in window {t_min} to {t_max} for spike {x}")
                pass
            
            
        #spike_count_list[i] = np.sum((spikes >= t_min) & (spikes < t_max)) <-- betere oplossing van chatgpt inplaats van nested for loop
    
    return spike_count_list
    

t = 400 * b2.ms 
t_window = 20 * b2.ms 
t_min = t - t_window / 2
t_max = t + t_window / 2


# calculate spike counts
spike_counts = get_spike_count(spike_monitor_excit, idx_monitored_neurons_excit, t_min, t_max)
print(spike_counts)

# convert spike counts to spike rates (in Hz)
spike_rates = spike_counts / (t_max - t_min)

plt.plot(spike_rates, ".b")
plt.xlabel("Neuron index")
plt.ylabel("Spike rate [Hz]")
plt.show() 

```

```python
def get_theta(t_min, t_max, spike_monitor, spike_index_list, N):
    """
    Computes the population vector (t).

    Parameters:
    - t_min, t_max: Start and end of the time window.
    - spike_monitor: Brian2 SpikeMonitor object.
    - spike_index_list: List of monitored neurons.
    - N: Total number of excitatory neurons.

    Returns:
    - Estimated (t) in degrees.
    """
    spike_counts = get_spike_count(spike_monitor, spike_index_list, t_min, t_max)
    orientations = get_orientation(spike_index_list, N)

    # compute population vector (weighted mean)
    if np.sum(spike_counts) > 0:
        theta_t = np.sum(spike_counts * orientations) / np.sum(spike_counts)
    else:
        theta_t = None
    
    return theta_t


t_min = 380 * b2.ms
t_max = 420 * b2.ms 

theta_t = get_theta(t_min, t_max, spike_monitor_excit, idx_monitored_neurons_excit, 1000)
print(f"Decoded  at t = 400ms: {theta_t} degrees")

```

```python
def get_theta_time_series(spike_monitor, spike_index_list, t_snapshots, t_window, N):
    """
    Computes (t) over multiple time snapshots.

    Parameters:
    - spike_monitor: Brian2 SpikeMonitor object.
    - spike_index_list: List of monitored neurons.
    - t_snapshots: Array of time points to compute (t).
    - t_window: Time window for averaging.
    - N: Total neurons in excitatory population.

    Returns:
    - A list of (t) values over time.
    """
    theta_ts = []
    half_window = t_window / 2  

    for t in t_snapshots:
        theta_t = get_theta(t - half_window, t + half_window, spike_monitor, spike_index_list, N)
        theta_ts.append(theta_t)
    
    return np.array(theta_ts)

import matplotlib.pyplot as plt

# define time snapshots and time window
t_window = 50 * b2.ms  # Choose an appropriate time window
snapshot_interval = 10 * b2.ms
t_snapshots = np.arange(300, 500, snapshot_interval/b2.ms) * b2.ms  # After stimulus ends

theta_ts = get_theta_time_series(spike_monitor_excit, idx_monitored_neurons_excit, t_snapshots, t_window, 1000)

plt.figure(figsize=(6,4))
plt.plot(t_snapshots/b2.ms, theta_ts, marker='o')
plt.xlabel("Time (ms)")
plt.ylabel("Decoded  (degrees)")
plt.grid(True)
plt.show()

```

We are working with a network of spiking neurons that stores a memory of a stimulus.
Each neuron has a preferred orientation (for example, one neuron prefers 0, another prefers 45, another prefers 90, etc.).
After stimulation, we want to decode which orientation the network is remembering at a given time (t).

Goal: Compute (t)

At time t, some neurons spike more than others.
The neurons that fire more often are more important in determining the remembered orientation.
We compute a weighted average of the preferred orientations, using spike counts as weights.
This gives us an estimate of (t); the angle currently stored in the network.



- ni = number of spikes from neuron i in a given time window
- i = preffered direction of neuron i
- sum runs over all monitored neurons

To compute (t), we need to do three things:

Step 1: Map Neuron Indices to Preferred Orientations

Each neuron in the network has a preferred orientation based on its index.
- Function: get_orientation(idx_list, N); Takes a list of neuron indices and maps them to their preferred orientations. Assumes orientations are evenly distributed between 0 and 360.

Step 2: Count Spikes in a Given Time Window

Now, we count how many times each neuron fired in a given time window [t_min, t_max]
- Function: get_spike_count(spike_monitor, spike_index_list, t_min, t_max) Takes a spike monitor (which records all spikes).
Counts how many times each neuron fired in a given time window.
Uses spike_monitor.spike_trains() to access spike times.

Step 3: Compute (t)

Now, we compute the population vector (the weighted mean of preferred orientations).
- Function: get_theta(t_min, t_max, spike_monitor, spike_index_list, N) Uses get_spike_count() to get spike counts. Uses get_orientation() to get preferred orientations. Computes (t) using a weighted mean formula.

Step 4: Compute (t) Over Time

```python
import numpy as np
import brian2 as b2
from neurodynex3.working_memory_network import wm_model
from neurodynex3.tools import plot_tools
import matplotlib.pyplot as plt

N_excitatory = 2048
N_inhibitory = 512
weight_scaling_factor = 1
num_trials = 3  

t_window = 50 * b2.ms  
snapshot_interval = 10 * b2.ms  
t_snapshots = np.arange(300, 500, snapshot_interval/b2.ms) * b2.ms  

def get_orientation(idx_list, N):
    return [round(((idx + 0.5) / N) * 360, 2) for idx in idx_list]

def get_spike_count(spike_monitor, spike_index_list, t_min, t_max):
    nr_neurons = len(spike_index_list)
    spike_count_list = np.zeros(nr_neurons)
    spike_trains = spike_monitor.spike_trains()
    for i, neuron_index in enumerate(spike_index_list): 
        spikes = spike_trains[neuron_index]
        spike_count_list[i] = np.sum((spikes >= t_min) & (spikes < t_max))  
    return spike_count_list

def get_theta(t_min, t_max, spike_monitor, spike_index_list, N):
    spike_counts = get_spike_count(spike_monitor, spike_index_list, t_min, t_max)
    orientations = get_orientation(spike_index_list, N)
    if np.sum(spike_counts) > 0:
        theta_t = np.sum(spike_counts * orientations) / np.sum(spike_counts)
    else:
        theta_t = None  
    return theta_t

def get_theta_time_series(spike_monitor, spike_index_list, t_snapshots, t_window, N):
    theta_ts = []
    half_window = t_window / 2
    for t in t_snapshots:
        theta_t = get_theta(t - half_window, t + half_window, spike_monitor, spike_index_list, N)
        theta_ts.append(theta_t)
    return np.array(theta_ts)

plt.figure(figsize=(8, 5))

for trial in range(num_trials):
    print(f"Running trial {trial + 1}...")

    rate_monitor_excit, spike_monitor_excit, voltage_monitor_excit, idx_monitored_neurons_excit, \
    rate_monitor_inhib, spike_monitor_inhib, voltage_monitor_inhib, idx_monitored_neurons_inhib, w_profile = wm_model.simulate_wm(
        stimulus_center_deg=120, 
        stimulus_width_deg=30, 
        stimulus_strength=.06 * b2.namp, 
        t_stimulus_start=100 * b2.ms, 
        t_stimulus_duration=200 * b2.ms, 
        sim_time=500. * b2.ms, 
        N_excitatory=N_excitatory,
        N_inhibitory=N_inhibitory, 
        weight_scaling_factor=weight_scaling_factor)

    theta_ts = get_theta_time_series(spike_monitor_excit, idx_monitored_neurons_excit, t_snapshots, t_window, N_excitatory)

    plt.plot(t_snapshots / b2.ms, theta_ts, marker='o', linestyle='-', label=f"Trial {trial + 1}")

plt.xlabel("Time (ms)")
plt.ylabel("Decoded  (degrees)")
plt.legend()
plt.grid(True)
plt.show()
```
